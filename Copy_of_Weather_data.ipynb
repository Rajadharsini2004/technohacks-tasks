```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/wether data.csv'
df = pd.read_csv(file_path)

print(type(df))
print(df.head())

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation of temperature over the cities')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/wether data.csv'
df = pd.read_csv(file_path)

df['datetime'] = pd.to_datetime(df['datetime'])

plt.figure(figsize=(10, 6))
plt.plot(df['datetime'], df['Los Angeles'], marker='o', linestyle='-', color='purple')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Los Angeles')
plt.grid(True)
plt.show()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.plot(df['datetime'], df['Indianapolis'], color='c')
plt.title('Temperature on Indianapolis Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.grid(True)
plt.show()
```
