```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("\nBasic statistics about the dataset:")
print(df.describe())
```

```python
data= pd.DataFrame(df)

X=df['Department']
Y=df['HourlyRate']

plt.bar(X, Y, color='pink')
plt.title("Employee department and working Hours")
plt.xlabel("Department of Employeers")
plt.ylabel("Working hours")

plt.show()
```

```python
x_values = df['MonthlyIncome']
y_values = df['JobInvolvement']
bubble_sizes = df['Education']

plt.scatter(x_values, y_values, s=bubble_sizes, alpha=0.5, cmap='viridis')
plt.xlabel('MonthlyIncome')
plt.ylabel('JobInvolvement')
plt.title('Bubble Chart')
plt.colorbar(label='Bubble Size')
plt.show()
```

```python
x_values = df['TotalWorkingYears']
y_values = df['MonthlyIncome']

plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, color='red', marker='o', alpha=0.7)

plt.xlabel('Working Years')
plt.ylabel('Monthly income')
plt.title('Monthly income of Employer based on their Experience')
plt.show()
```

```python
values = df['PercentSalaryHike']

plt.figure(figsize=(10, 6))
plt.hist(values, bins=10, color='blue', edgecolor='black', alpha=0.7)

plt.xlabel('Employees')
plt.ylabel('Percentage salary hike')
plt.title('Histogram')

plt.show()
```

```python
import numpy as np
```

```python
selected_column = 'HourlyRate'

column_data = df[selected_column]

mean = column_data.mean()
std_dev = column_data.std()
sample_size = len(column_data)

data_normal_dist = np.random.normal(mean, std_dev, sample_size)

plt.figure(figsize=(10, 6))
plt.hist(column_data, bins=30, density=True, color='blue', edgecolor='black', alpha=0.7, label='Actual Data')
plt.hist(data_normal_dist, bins=30, density=True, color='orange', edgecolor='black', alpha=0.5, label='Normal Distribution')

plt.xlabel('Hourly Rating')
plt.ylabel('Frequency')
plt.title('Normal Distribution Histogram')

plt.legend()

plt.show()
```

```python
x=df['YearsAtCompany']
y=df['Age']
plt.plot(x,y)
plt.xlabel('YearsAtCompany')
plt.ylabel('Age')
plt.title('Working Experience in their company based on their age')
plt.show()
```
